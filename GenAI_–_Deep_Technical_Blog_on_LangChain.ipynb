```python
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

pipe = pipeline("text-generation", model="distilgpt2", max_new_tokens=50)

llm = HuggingFacePipeline(pipeline=pipe)

print(llm.invoke("What is LangChain?").strip())
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template("Explain {topic} in simple terms")

print(prompt.format(topic="AI"))
```

```python
from langchain_core.output_parsers import StrOutputParser

chain = prompt | llm | StrOutputParser()

print(chain.invoke({"topic": "Machine Learning"}).strip())
```

```python
chat_history = []

user1 = "Hi"
chat_history.append(user1)

response1 = llm.invoke(user1)
chat_history.append(response1)

print("AI:", response1.strip())

user2 = "What did I just say?"
chat_history.append(user2)

context = " ".join(chat_history)

response2 = llm.invoke(context)

print("AI:", response2.strip())
```
